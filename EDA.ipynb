# EDA (EXPLORATORY DATA ANALYSIS)

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  2. Load Titanic Train Dataset

```python
file=pd.read_csv('titanic_train.csv')
file.head(10)
```

## 3. To Know The Values Are Null Or Not

```python
file.isnull()
```

## 4. To Know How Many Values Are Null Values In Each Columns 

```python
count_of_missing_values=file.isnull().sum()
count_of_missing_values[0:12]
```

## 5. To Know What Datatype It Is

```python
file.Name.dtype
```

```python
file.Pclass.dtype
```

## 6. To Fill NaN Values With Some Other Value 

```python
file.fillna('not found')
```

##  7. To replace anything in columns

```python
file.Embarked.replace('S','B')
```

## 8. Line Plot

```python
plt.figure(figsize=(14,6))
plt.title("titanic data in lineplot")
sns.lineplot(data=file,x='Age',y='Survived')
plt.show()
```

## 9. Bar Plot

```python
sns.barplot(data=file,x="Sex",y="Pclass")
```

## 10. Heatmap

```python
plt.figure(figsize=(14,6))
sns.heatmap(file.isnull())
```

## 11. Distplot

```python
sns.distplot(file['Age'].dropna(),color='blue',bins=10)
```

## 12. Boxplot

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='Pclass',y='Embarked',data=file,palette='Dark2_r')
```

## 13. Scatterplot

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x='Embarked',y='Age',data=file,palette='Dark2_r')
```

## 14. Pairplot

```python
plt.figure(figsize=(14,10))
sns.pairplot(file)
```

## 15. Cleaning Dataset

```python
file.drop('Name',axis=1,inplace=True)
```
